# 9.2.1 Scopes and Namespaces Example

```python
def scope_test():
    def do_local():
        spam = "local spam" #작은 def 안에서 
    
    def do_nonlocal():
        nonlocal spam #실행되는 scope_test()
        spam = "nonlocal spam"
        
    def do_global():
        global spam #scope_test()안에서는 변경안된다. 
        spam = "global spam"
    
    spam = "test spam"
    do_local()
    print("After local assignment:",spam)
    do_nonlocal()
    print("After nonlocal assignment:",spam)
    do_global()
    print("After global assignment:",spam)
    
    
scope_test()
print("In global scope:",spam)
```

```python
class Rectangle:
    count = 0
    
    def __init__(self, width, height):
        self.width = width
        self.height = height
        Rectangle.count += 1
    
    #인스턴스 메서드
    def calcArea(self):
        area  = self.width * self.height
        return area
    
    #정적 메서드
    @staticmethod
    def isSqare(rectWidth, rectHeight):
        return rectWidth == rectHeight
    
    #클래스 메서드
    @classmethod
    def printCount(cls):
        print(cls.count)
```

```python
square = Rectangle.isSqare(5, 5)
print(square)
```

```python
rect1 = Rectangle(5, 5)
rect2 = Rectangle(2, 5)
rect1.printCount()
```

```python
def __add__(self, other):
    obj = Rectangle(self.width + other.width, self.height + other.height)
    return obj

r3 = 1 + 2
print(r3)
```

### 4. 클래스 상속과 다형성

```python
class Animal:
    def __init__(self, name):
        self.name = name
    def move(self):
        print("move")
    def speak(self):
        pass
    
    
class Dog (Animal):
    def speak(self):
        print("bark")

        
class Duck(Animal):
    def speak(self):
        print("quack")
```

```python
dog = Dog("doggy")
n = dog.name
dog.move()
dog.speak()
```

```python
animals = [Dog('doggy'), Duck('duck')]

for i in animals:
    i.speak()
```
