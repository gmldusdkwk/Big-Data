## 1. 이미지 url를 뽑아낸다.

```python
def getURL(url):
    html = requests.get(url)
    soup = BeautifulSoup(html.text, "lxml")
    img = soup.find_all("img")
    
    return img
```

```python
data = getURL('https://pixabay.com/ko/')
```

```python
data
```

## 2. 그 중 http://로 시작하는 것만 리스트에 저장한다. 

```python
import re
result = []
count = 0
```

```python
for row in data:
    #print(row.get('src'))
    m = re.search(r'^(https?:\/\/)[a-zA-Z.\/0-9\-\__]{1,}',row.get('src'))
    try:
        count += 1
        print(m.group(0))
        result.append([count,m.group(0)])
    except AttributeError:
        print(AttributeError)
```

```python
for i in result:
    print(i)
print(len(result))
```
