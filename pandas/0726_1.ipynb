## filter

2. like
3. regex: regular expression

```python
data.filter(items=['State', 'Metro'])
```

```python
data.filter(regex='^Region', axis=0)
```

```python
data['Zhvi'] > 500000  # 결과가 True, False 로 나온다. 그래서 loc 안 조건으로 넣으면 된다. 
```

```python
data['Zhvi']
```
